```python
preds=[]
vals=[]
with open('./Results/summarization3_results.txt','r') as fd:
    while(True):
        line=fd.readline()
        if(line==''):
            break
        if("y_pred:" in line):
            preds.append(line.strip().split(':')[1].strip())
        else:
            vals.append(line.strip().split(':')[1].strip())
```

```python

#this is an implementation of unigram bleu score.
def unigram_bleu_score(pred_list,val_list):
    score=0
    N=0
    for i in range(len(pred_list)):
        pred=pred_list[i].split()
        
        val=val_list[i].split()
       
        c_pred=Counter(pred)
        c_val=Counter(val)
        N+=len(val)
        for key,value in c_pred.items():
            score+=min(value,c_val[key])
    print("Score: ",score,"N: ",N)        
    bleu_score=score/N
    return bleu_score

def bigram_bleu_score(pred_list,val_list):
          score=0
          N=0
          for i in range(len(pred_list)):
              pred=pred_list[i].split()
              val=val_list[i].split()
            
              pred=list(zip(pred[:-1],pred[1:]))
              val=list(zip(val[:-1],val[1:]))
              c_pred=Counter(pred)
              c_val=Counter(val)
              N+=len(val)
              for key,value in c_pred.items():
                  score+=min(value,c_val[key])
          print("Score: ",score,"N: ",N)
          bleu_score=score/N
          return bleu_score        
```

```python
bleu_score=unigram_bleu_score(preds,vals)
print(bleu_score)
bleu_score=bigram_bleu_score(preds,vals)
print(bleu_score)
```

```python
#plot the results in scatter plot using bokeh.
#the circle sizes depend on how many matches in each sentence.
idxs=list(range(len(preds)))
match_scores=[]
val_lengths=[]
pred_lengths=[]
def similarity(pred_list,val_list):
    global match_scores
    for i in range(len(pred_list)):
        score=0
        pred=pred_list[i].split()
        val=val_list[i].split()
        c_pred=Counter(pred)
        c_val=Counter(val)
        N=len(val)
        val_lengths.append(len(val))
        pred_lengths.append(len(pred))
        for key,value in c_pred.items():
            score+=min(value,c_val[key])
            
        match_scores.append(score/N)    
            
similarity(preds,vals)  
df=pd.DataFrame(dict(idxs=idxs,preds=preds,vals=vals,scores=match_scores,\
                                    val_lengths=val_lengths,pred_lengths=pred_lengths))

source_summary=ColumnDataSource(data=dict(idxs=idxs,preds=preds,vals=vals,scores=match_scores,\
                                    val_lengths=val_lengths,pred_lengths=pred_lengths))

plot=None
slider=None
slider2=None
def update_plot(attr,old,new):
    global plot,slider,df,source_summary
    percent=slider.value
    filtered=df.loc[df['scores']>=percent,:]
    print(filtered.head())
    source_summary.data['idxs']=filtered['idxs']
    source_summary.data['preds']=filtered['preds']
    source_summary.data['vals']=filtered['vals']
    source_summary.data['scores']=filtered['scores']
    source_summary.data['val_lengths']=filtered['val_lengths']
    source_summary.data['pred_lengths']=filtered['pred_lengths']
    
def update_plot(attr,old,new):
    global plot,slider2,df,source_summary
    val_length=slider2.value
    filtered=df.loc[df['val_lengths']==val_length,:]
    print(filtered.head())
    source_summary.data['idxs']=filtered['idxs']
    source_summary.data['preds']=filtered['preds']
    source_summary.data['vals']=filtered['vals']
    source_summary.data['scores']=filtered['scores']
    source_summary.data['val_lengths']=filtered['val_lengths']
    source_summary.data['pred_lengths']=filtered['pred_lengths']
    
slider = Slider(start=0, end=1, value=1, step=0.1, title="Unigram bleu scores")
slider.on_change('value', update_plot)

slider2 = Slider(start=1, end=15, value=1, step=1, title="Val lengths")
slider2.on_change('value', update_plot2)

size_mapper=LinearInterpolator(x=[min(source_summary.data['scores']),max(source_summary.data['scores'])],\
y=[5,30])

color_mapper=LinearColorMapper(low=min(source_summary.data['pred_lengths']),\
                               high=max(source_summary.data['pred_lengths']),\
                              palette=viridis(256))

hover=HoverTool(tooltips=[('y_pred:',"@preds"),\
                          ('y_val:','@vals'),\
                           ('Idx: ','@idxs'),\
                           ('Score: ','@scores') ],show_arrow=False)

plot=figure(plot_width=1000, plot_height=700,x_axis_label='Val points',y_axis_label='Val lengths',\
                )
plot.circle('idxs','val_lengths',\
            size={'field':'scores', 'transform': size_mapper},\
            color={'field': 'pred_lengths', 'transform': color_mapper},\
            source=source_summary,alpha=0.1)
    plot.add_tools(hover)
layout=column(row(widgetbox(slider),widgetbox(slider2)),plot)
curdoc().add_root(layout)
show(plot)
```
